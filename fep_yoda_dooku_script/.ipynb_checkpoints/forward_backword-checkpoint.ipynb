```python
data=pd.read_table('./open_dooku.txt',
                     header=None,delim_whitespace=True)
data.columns=['time','fep']

from scipy.integrate import trapz
data['dG']=data['fep'].diff()
data['dG'][0]=0
data['dt']=data['time']-np.min(data['time'])
data_f=integrate.cumtrapz(data.fep,data.dt,initial=0)
print data
print data_f
dataBack=data.iloc[::-1]

dataBack['dG']=dataBack['fep'].diff()
dataBack['dG'][4]=0
print dataBack
data_b=integrate.cumtrapz(dataBack.fep,data.dt,initial=0)
print data_b

plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')
from matplotlib.pyplot import figure
plt.figure(figsize=(12,4))
#plt.plot(data.time,data.fep,'y-',label='forward')
plt.plot((data.time-min(data.time))/(max(data.time)-min(data.time)),data_f,'y-',label='forward',color='black')
plt.plot((data.time-min(data.time))/(max(data.time)-min(data.time)),data_b,'y-',label='backward',color='red')
plt.legend(fontsize=10)
plt.title('dooku-open',fontsize=30)
plt.xlabel(r"Fraction of Time",color="black",fontsize=30)
plt.ylabel(r"dG",color="black",fontsize=30)
plt.tick_params(axis='x', colors='black')
plt.tick_params(axis='y', colors='black')
plt.tick_params(axis='x', labelsize=25)
plt.tick_params(axis='y', labelsize=25)
```
